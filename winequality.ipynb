```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
#import random forest
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
dataset=pd.read_csv('/content/winequality-red.csv')
```

```python
dataset.head()
```

```python
dataset['quality'].unique()
```

```python
#check null values
dataset.isnull().sum()
```

wine quality: less if <5

wine quality :med if=5

wine quality : good if>5

```python
#sine wine quality has these values[5, 6, 7, 4, 8, 3] , we will  reclassify  wine quality labels into three categories:

#Bad (0): Quality < 5
#Fine (1): Quality = 5
#Good (2): Quality > 5
def redefine_quality(quality):
  if quality<5:
    return 0 #bad
  elif quality==5:
    return 1 #medium
  else:
    return 2  #good
dataset['Quality_Label'] = dataset['quality'].apply(redefine_quality)
```

```python
dataset.head()
```

```python
#seperate labels and data
X=dataset.iloc[:,:-1]
Y=dataset.iloc[:,-1]
print(Y)
```

```python
#train the model
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
# Train Random Forest Classifier
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train, Y_train)

# Predictions
Y_pred = rf_model.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(Y_test, Y_pred)
print(f'Accuracy: {accuracy:.2f}')
```

```python


# Define input data (example values)
input_data = (7.8, 0.7, 0.56, 1.9, 0.076, 25.0, 34, 0.9971, 3.50, 0.5, 9.4, 5.0)  # Exclude the last value (quality)

# Convert input data into a NumPy array
input_array = np.asarray(input_data).reshape(1, -1)  # Reshape to match model input shape

# Make a prediction
prediction = rf_model.predict(input_array)

# Output result
if prediction[0] == 0:
    print('Bad Wine')
elif prediction[0] == 1:
    print('Fine Wine')
else:
    print('Good Wine')
```
