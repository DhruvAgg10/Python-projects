The problem is of binary classification because it has only 2 fields: whether its a rock or a bomb

WORKFLOW: import data-> preprocess data -> split data into test and train -> apply logistic regression as machine learning algo to train the data.

Step 1- import all the dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
from sklearn.ensemble import RandomForestClassifier
#trying to improve model accuracy
#from sklearn.preprocessing import MinMaxScaler
#this wont improve acc because our data is already scaled between 0-1
```

Step2: import and preprocess data

```python
dataset= pd.read_csv('/content/Copy of sonar data.csv', header=None)
#as there is no header file in our dataset therefore we need to mention it
```

```python
dataset.shape
```

```python
dataset[60].value_counts()
# its a rock is given in 60th column there fore put dataset[60] to count value
```

Now we have to seperate the data and the labels

```python
# drop the labels i.e the 60th column from dataset and put it in another variable
#X= dataset.drop(columns=60, axis=1)
#Y=dataset[60]
X = dataset.iloc[:, :-1]  # Assuming last column is the target (mine or rock)
Y = dataset.iloc[:, -1]    # Labels (binary classification: mine=0, rock=1)
#both do the same function but diff is that in above code the last col is considerd as 60th only so if we
#insert anything before it it will still take only 60th col, but in 2nd code it automatically takes
#the last col even if we add something before it

```

Now split the data into train and test

```python
X_train, X_test, Y_train, Y_test= train_test_split(X, Y, test_size=0.5, stratify=Y, random_state=1)
#test_size= 0.1 means 10% data will be used for testing
# stratify=Y : Purpose: Ensures that the distribution of classes
 #(or target variable values) in the original dataset (Y) is preserved
 #in both the training (Y_train) and testing (Y_test) sets.
# random_state=1: ensures that every time we run the code we get same test and train data
```

Train the data on logistic regression model

```python
model=LogisticRegression()
```

```python
# Create a Random Forest classifier with 100 trees
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)

# Train the model
rf_model.fit(X_train, Y_train)
```

```python
# train the model using training data
model.fit(X_train, Y_train)
```

Now evaluate the model

```python
X_train_prediction= rf_model.predict(X_train)
training_data_accuracy= accuracy_score(X_train_prediction, Y_train)
print('accuracy', training_data_accuracy)
```

```python
#accuracy on trainning data
X_train_prediction= model.predict(X_train)
training_data_accuracy= accuracy_score(X_train_prediction, Y_train)
# Y_train is the real ans and X_train_prediction is the predicted answer
```

```python
print('accuracy', training_data_accuracy)
# accuracy using just logistic regression came out to be 86%
```

```python
X_test_prediction= rf_model.predict(X_test)
training_data_accuracy= accuracy_score(X_test_prediction, Y_test)
print('accuracy', training_data_accuracy)
```

```python
# now test the accuracy of test data
X_test_prediction= model.predict(X_test)
test_data_accuracy= accuracy_score(X_test_prediction, Y_test)
```

```python
print('accuracy', test_data_accuracy)
## accuracy using just logistic regression came out to be 77%
#so random forest is better than logistic regression
```
